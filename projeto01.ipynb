```python
class CreditCard:
    def __init__(self, customer, bank, acnt, limit):
        self._customer = customer
        self._bank = bank
        self._account = acnt
        self._limit = limit
        self._balance = 0

    def get_customer(self):
        return self._customer

    def get_bank(self):
        return self._bank

    def get_account(self):
        return self._account

    def get_limit(self):
        return self._limit

    def get_balance(self):
        return self._balance

    def charge(self, price):
        if price + self._balance > self._limit:
            return False
        else:
            self._balance += price
            return True

    def make_payment(self, amount):
        self._balance -= amount


def main():
    credit_cards = {}

    while True:
        print("1. Cadastrar novo cartão")
        print("2. Selecionar cartão")
        print("3. Efetuar compra")
        print("4. Fazer pagamento")
        print("5. Sair")

        choice = input("Digite a opção desejada: ")

        if choice == "1":
            customer = input("Nome do cliente: ")
            bank = input("Nome do banco: ")
            acnt = input("Número da conta: ")
            limit = float(input("Limite do cartão: ")) 
            credit_cards[acnt] = CreditCard(customer, bank, acnt, limit)
            print("Cartão cadastrado com sucesso!")

        elif choice == "2":
            acnt = input("Digite o número da conta: ")
            if acnt in credit_cards:
                selected_card = credit_cards[acnt]
                print(f"Cartão selecionado: {selected_card.get_customer()} - {selected_card.get_bank()} - {selected_card.get_account()}")
            else:
                print("Cartão não encontrado.")

        elif choice == "3":
            acnt = input("Digite o número da conta: ")
            if acnt in credit_cards:
                selected_card = credit_cards[acnt]
                price = float(input("Digite o valor da compra: "))
                if selected_card.charge(price):
                    print("Compra efetuada com sucesso!")
                else:
                    print("Limite de crédito excedido.")
            else:
                print("Cartão não encontrado.")

        elif choice == "4":
            acnt = input("Digite o número da conta: ")
            if acnt in credit_cards:
                selected_card = credit_cards[acnt]
                amount = float(input("Digite o valor do pagamento: "))
                selected_card.make_payment(amount)
                print("Pagamento efetuado com sucesso!")
            else:
                print("Cartão não encontrado.")

        elif choice == "5":
            print("Saindo...")
            break

        else:
            print("Opção inválida. Por favor, escolha novamente.")

main()
```
